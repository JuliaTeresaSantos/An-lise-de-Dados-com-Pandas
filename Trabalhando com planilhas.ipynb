# **Planilhas** **do** **Excel**




```python
#importar a biblioteca
import pandas as pd 
```

```python
#ler os arquivos excel
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo as 5 Ãºltimas linhas
df.tail()
```

```python
#Exibe uma amostra do dataset
df.sample(5)
```

```python
#Saber qual o tipo de dado de cada coluna
df.dtypes
```

```python
#Alterar o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.head()
```

# **Valores** **Faltantes** - tratamento dos dados

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela mÃ©dia
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

**Para** **substituir** **os** **valores** **nulo** **por** **0** 

df["Vendas"].fillna(0, inplace=True)

**Para** **apagar** **as** **linhas** **com** **valores** **nulos** **se** **baseando** **em** **1** **coluna**

df.dropna(subset=["Vendas"], inplace=True)

**Para** **apagar** **as** **linhas** **com** **valores** **nulos** **se** **baseando** **em** **todas** **as** **colunas**

df.dropna(how="all", inplace=True)

# **Adicionando** **colunas** **no** **dataset**

```python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#adicionando uma coluna que calcula a receita/vendas (qtde)  
df["Receita/Vendas"] = df["Receita"] / df["Vendas"] 
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest -> 3 melhores (maiores) receitas
df.nlargest(3, "Receita")
```

```python
#nsamllest -> mede as 3 piores (menores) receitas
df.nsmallest(3, "Receita")
```

**Agrupamento**, **ordenando**

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

# **Datas**

```python
#Transformando a coluna de data em tipo inteiro (simulando um cenÃ¡rio real das empresas)
df["Data"] = df["Data"].astype("int64")
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mÃªs e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferenÃ§a de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mÃªs de marÃ§o
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19.sample(20)
```
